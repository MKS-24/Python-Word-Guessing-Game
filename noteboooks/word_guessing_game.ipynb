```python
import streamlit as st
import random

# --- Page Setup ---
st.set_page_config(page_title="Word Guessing Game", layout="centered")

st.title("Welcome To The Word Guessing Game")

# --- Initialize session state ---
if "game_started" not in st.session_state:
    st.session_state.game_started = False
if "game_over" not in st.session_state:
    st.session_state.game_over = False
if "save" not in st.session_state:
    st.session_state.save = None
if "question" not in st.session_state:
    st.session_state.question = ["cat", "dog", "hen", "goat", "cow", "frog", "snake", "turtle"]

# --- Start or Exit Menu ---
if not st.session_state.game_started:
    choice = st.radio("What Do You Want?", ["Start The Game", "Exit From Game"])
    if st.button("Confirm"):
        if choice == "Start The Game":
            st.session_state.game_started = True
            st.session_state.game_over = False
            st.session_state.save = str(random.choice(st.session_state.question))
        else:
            st.success("ðŸ‘‹ Allah Hafiz")
            st.stop()

# --- Main Game ---
if st.session_state.game_started and not st.session_state.game_over:
    st.write("### Choose from the following options:")

    # Display numbered list (0-based index like your example)
    for i, q in enumerate(st.session_state.question):
        st.write(f"{i}: {q}")

    answer = st.text_input("ðŸ”¤ Enter your guess (word or number):")

    if st.button("Submit Answer"):
        if not answer.strip():
            st.warning("Please enter a value before submitting.")
        else:
            # Prepare string versions for comparison
            question_str = [str(x) for x in st.session_state.question]
            correct_value = str(st.session_state.save)

            # If user entered a valid index number, map to that value
            if answer.isdigit() and int(answer) < len(question_str):
                user_value = question_str[int(answer)]
            else:
                user_value = answer

            # Case-insensitive check if user's input exists in options
            match_index = None
            for idx, val in enumerate(question_str):
                if user_value.lower() == val.lower():
                    match_index = idx
                    break

            if match_index is not None:
                # Find correct answer index
                correct_index = next(
                    (i for i, v in enumerate(question_str) if v.lower() == correct_value.lower()),
                    None
                )

                if user_value.lower() == correct_value.lower():
                    st.balloons()
                    st.success(
                        f"ðŸŽ‰ Congrats! You guessed correctly!\n\nOption {correct_index}: **{correct_value}**"
                    )
                    st.session_state.game_over = True  # Now show Play Again / Exit after correct guess
                else:
                    if correct_index is not None:
                        st.error("âŒ Wrong guess!")
                        st.info(f"Correct answer was ðŸ‘‰ Option {correct_index}: **{question_str[correct_index]}**")
                    else:
                        st.error(f"âŒ Wrong guess. The correct answer was {correct_value}.")
                    st.warning("---- Game Over ----")
                    st.session_state.game_over = True
            else:
                st.warning("Please enter a valid option (either number or word from the list).")

# --- Play Again / Exit Section ---
if st.session_state.game_started and st.session_state.game_over:
    col1, col2 = st.columns(2)
    with col1:
        if st.button("ðŸ” Play Again"):
            st.session_state.save = str(random.choice(st.session_state.question))
            st.session_state.game_over = False
    with col2:
        if st.button("ðŸšª Exit"):
            st.success("ðŸ‘‹ Allah Hafiz")
            st.session_state.game_started = False
elif st.session_state.game_started and not st.session_state.game_over:
    if st.button("ðŸšª Exit Game"):
        st.success("ðŸ‘‹ Allah Hafiz")
        st.session_state.game_started = False
```
